#Label Encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

**Label Encoding**


*   Converting the labels into numeric form



```python
cancer_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/Learn from outside/type 1/Copy of breast_cancer_data.csv')
```

```python
cancer_data.head()
```

```python
#check the data
cancer_data['diagnosis'].value_counts()
```

**convert text data into numeric data**

```python
# load LabelEncoder
label_encoder = LabelEncoder()
```

```python
label = label_encoder.fit_transform(cancer_data.diagnosis)
label
```

**appending the label into dataframe**

```python
cancer_data ['target'] = label
```

```python
cancer_data.head()
```

**In target column**


0-->Benign


1-->Malignant

```python
cancer_data['target'].value_counts()
```
