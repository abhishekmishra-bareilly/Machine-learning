# Imbalanced dataset using credit card dataset

```python
import numpy as np
import pandas as pd
```

```python
# Load the data
credit_card_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/Learn from outside/type 1/Copy of credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data['Class'].value_counts()
```

**So this is very imbalanced dataset**

**columns discription**

Time--> How much time taken for the transtion

Amount--> How much amount transfferd us doller

Class--> 
          
         0 - ligil transtion
         1 - froud transtion  

```python
ligil= credit_card_data[credit_card_data.Class==0]
froud = credit_card_data[credit_card_data.Class==1]
```

```python
ligil.shape
```

```python
froud.shape
```

**Using under sampling for Imbalanced dataset**

```python
ligil_sampling = ligil.sample(n = 492)
```

```python
print(ligil_sampling.shape)
```

**Now we are going to concat both data "ligil_sampling", "froud"**

```python
new_dataset = pd.concat((ligil_sampling,froud),axis = 0)
new_dataset
```
