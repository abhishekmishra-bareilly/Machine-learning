```python
 # importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

**We will be working on the breast cancer dataset**

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python

print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python

# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# number of rows and Columns in this dataset
data_frame.shape
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# checking the distribution of Target Varibale
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant



### **Separating the features and target**

```python
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x = np.asarray(x)
y = np.asarray(y)
```

## **GridSearchCV**

**GridSearchCV is used for determining the best parameters for our model**

```python
# loading model
model = SVC()
```

```python
# Hyperparameters

perameters =  {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
# Grid search

classifier = GridSearchCV(model, perameters, cv = 5)
```

```python
classifier.fit(x, y)
```

```python
# fitting the data to our model
classifier.cv_results_
```

```python
#  Best parameters
best_perameters =  classifier.best_params_
print(best_perameters)
```

```python
# high Accuracy 
best_score=classifier.best_score_
print(best_score)
```

```python
# Loding the result into pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_result = result[['param_C','param_kernel','mean_test_score']]
grid_result
```

**Highest Accuracy = 95.2%**

**Best Parameters = {'C':10, 'kernel':'linear'}**

## **RandomizedSearchCV**

```python
# loading model
model = SVC()
```

```python
# Hyperparameters

perameters =  {
              'kernel':['linear','poly','rbf','sigmoid'],
              'C':[1, 5, 10, 20]
}
```

```python
#RandomizedSearchCV search

classifier = RandomizedSearchCV(model, perameters, cv = 5)
```

```python
classifier.fit(x, y)
```

```python
# fitting the data to our model
classifier.cv_results_
```

```python
#  Best parameters
best_perameters =  classifier.best_params_
print(best_perameters)
```

```python
# high Accuracy 
best_score=classifier.best_score_
print(best_score)
```

```python
# Loding the result into pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_result = result[['param_C','param_kernel','mean_test_score']]
grid_result
```

**Highest Accuracy = 95.0%**

**Best Parameters = {'C':5, 'kernel':'linear'}**
