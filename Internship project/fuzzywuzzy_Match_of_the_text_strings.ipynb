```python
# we try to split the columns 
column_a, column_s = string_matching['Title'], string_matching['List of Causes']
```

**Using fuzzywuzzy function**

```python
# Creating a function using fuzzywuzzy process
def get_matches(q,choices,limit=254):
  result = process.extract(q,choices,limit=limit)
  return result
```

**Now we can find the string maching % for each row**

```python
# String maching for first row of column A.
get_matches(column_a[1],column_s)
```
